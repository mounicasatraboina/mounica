```python
# Tagging: Tag Parts of Speech of each word in the tex
import pandas as pd

dataframe = pd.read_csv('output.csv')
data = dataframe['Review text']
data

from nltk.tokenize import word_tokenize
pos = []
for statement in data:
  text = word_tokenize(statement)
  pos.append(nltk.pos_tag(text))
pos
```

```python
from collections import Counter
count = []
for tags in pos:
  counts = Counter(tag for words, tag in tags)
  count.append(counts)
count
```
